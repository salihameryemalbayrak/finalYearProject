Define a function to load sismogram data and labels from the HDF5 file, ensuring consistent dimensions by padding or cropping, and apply standardization. This function will also handle loading potential metadata if needed later. Convert the loaded data and labels to NumPy arrays.



```python
import h5py
import numpy as np
import os

def load_and_preprocess_data(file_path, sismogram_length=540):
    """
    Loads sismogram data and labels from HDF5 file, preprocesses them.

    Args:
        file_path (str): Path to the HDF5 file.
        sismogram_length (int): Desired length of sismograms after padding/cropping.

    Returns:
        tuple: (sismogram_data, labels, metadata) where:
            sismogram_data (np.ndarray): Preprocessed sismogram data.
            labels (np.ndarray): Corresponding labels (0 for AN, 1 for EQ).
            metadata (dict): Dictionary containing metadata (e.g., station info).
                             Currently loads station info, but can be extended.
    """
    sismogram_data = []
    labels = []
    metadata = {}

    if not os.path.exists(file_path):
        print(f"File not found at {file_path}")
        return None, None, None

    with h5py.File(file_path, 'r') as f:
        # Load AN data (label 0)
        if 'AN' in f and isinstance(f['AN'], h5py.Group):
            print("Loading AN data...")
            for key in f['AN'].keys():
                dataset = f['AN'][key]
                if isinstance(dataset, h5py.Dataset):
                    data = dataset[:]
                    # Ensure consistent dimensions (padding or cropping)
                    if data.shape[1] < sismogram_length:
                        # Pad with zeros
                        padding_width = sismogram_length - data.shape[1]
                        data = np.pad(data, ((0, 0), (0, padding_width)), mode='constant')
                    elif data.shape[1] > sismogram_length:
                        # Crop
                        data = data[:, :sismogram_length]

                    # Standardization (mean 0, std dev 1)
                    mean = np.mean(data, axis=1, keepdims=True)
                    std = np.std(data, axis=1, keepdims=True)
                    # Avoid division by zero
                    std[std == 0] = 1e-8
                    data = (data - mean) / std

                    sismogram_data.append(data)
                    labels.append(0)
                else:
                    print(f"Skipping non-dataset member in AN: {key}")


        # Load EQ data (label 1)
        if 'EQ' in f and isinstance(f['EQ'], h5py.Group):
            print("Loading EQ data...")
            for key in f['EQ'].keys():
                dataset = f['EQ'][key]
                if isinstance(dataset, h5py.Dataset):
                    data = dataset[:]
                     # Ensure consistent dimensions (padding or cropping)
                    if data.shape[1] < sismogram_length:
                        # Pad with zeros
                        padding_width = sismogram_length - data.shape[1]
                        data = np.pad(data, ((0, 0), (0, padding_width)), mode='constant')
                    elif data.shape[1] > sismogram_length:
                        # Crop
                        data = data[:, :sismogram_length]

                    # Standardization (mean 0, std dev 1)
                    mean = np.mean(data, axis=1, keepdims=True)
                    std = np.std(data, axis=1, keepdims=True)
                     # Avoid division by zero
                    std[std == 0] = 1e-8
                    data = (data - mean) / std

                    sismogram_data.append(data)
                    labels.append(1)
                else:
                    print(f"Skipping non-dataset member in EQ: {key}")

        # Load Stations metadata (example)
        if 'Stations' in f and isinstance(f['Stations'], h5py.Group):
            print("Loading Stations metadata...")
            station_info = {}
            for key in f['Stations'].keys():
                 dataset = f['Stations'][key]
                 if isinstance(dataset, h5py.Dataset):
                     # Assuming station data is small and can be loaded directly
                     try:
                         station_data = dataset[:]
                         # Decode bytes if necessary
                         if dataset.dtype.kind == 'S':
                              station_info[key] = [item.decode('utf-8') for item in station_data]
                         else:
                              station_info[key] = station_data
                     except Exception as e:
                         print(f"Error loading station metadata for {key}: {e}")
            metadata['Stations'] = station_info
        else:
            print("'Stations' group not found or not a group.")


    # Convert lists to NumPy arrays
    sismogram_data_np = np.array(sismogram_data)
    labels_np = np.array(labels)

    print(f"Loaded {sismogram_data_np.shape[0]} sismograms with shape {sismogram_data_np.shape[1:]}")
    print(f"Loaded {labels_np.shape[0]} labels.")
    if 'Stations' in metadata:
        print(f"Loaded metadata for {len(metadata['Stations'])} stations.")


    return sismogram_data_np, labels_np, metadata

# Define file path
file_path = '/content/drive/MyDrive/Bitirme/LEN-DB-mini.hdf5'

# Load and preprocess the data
sismogram_data, labels, metadata = load_and_preprocess_data(file_path)

# Display shapes to verify
if sismogram_data is not None and labels is not None:
    print("\nProcessed Data Shapes:")
    print("Sismogram Data Shape:", sismogram_data.shape)
    print("Labels Shape:", labels.shape)
    if metadata and 'Stations' in metadata:
        print("Stations Metadata Keys:", list(metadata['Stations'].keys())[:5]) # Display first 5 keys
```

## Model mimarisini tanımlama

### Subtask:
Sismogram verilerini sınıflandırmak için uygun bir makine öğrenimi modeli mimarisi seçin ve tanımlayın. Araştırma makalesindeki bilgileri ve veri setinin yapısını göz önünde bulundurun.


**Reasoning**:
Based on the previous data exploration and the task requirements, the next step is to define and compile a CNN model suitable for classifying the loaded sismogram data. This involves defining the input shape based on the preprocessed data (number of channels, time steps), adding appropriate layers like Conv1D and MaxPooling1D for feature extraction, flattening the output for dense layers, and using a final dense layer with a sigmoid activation for binary classification. The model will then be compiled with binary crossentropy loss and the Adam optimizer.



```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam

# Assuming sismogram_data shape is (num_samples, num_channels, time_steps)
# and we want to treat channels as separate features initially or use Conv1D across time
# Let's define the input shape as (time_steps, num_channels) by transposing
# or (num_channels, time_steps) and use Conv1D on time dimension.
# Given the shape (186743, 3, 540), it's likely (samples, channels, time_steps)
# For Conv1D, the input shape should be (time_steps, features).
# We can either process each channel separately or treat channels as features for each time step.
# Let's try treating channels as features for each time step (shape (time_steps, channels)).
# This requires reshaping the input data.

# Alternatively, a common approach for multi-channel time series is to use Conv1D with input_shape=(time_steps, num_channels).
# Let's stick to this more standard approach, where the convolution happens along the time axis
# across all channels simultaneously.

# Input shape: (time_steps, num_channels)
input_shape = (sismogram_data.shape[2], sismogram_data.shape[1]) # (540, 3)

# Define the model architecture
model = Sequential()

# Convolutional Layer 1
model.add(Conv1D(filters=32, kernel_size=5, activation='relu', input_shape=input_shape))
model.add(BatchNormalization())
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.25))

# Convolutional Layer 2
model.add(Conv1D(filters=64, kernel_size=5, activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.25))

# Convolutional Layer 3
model.add(Conv1D(filters=128, kernel_size=5, activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.25))

# Flatten the output of the convolutional layers
model.add(Flatten())

# Dense Layer 1
model.add(Dense(128, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))

# Output Layer for binary classification
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Print the model summary
model.summary()
```
